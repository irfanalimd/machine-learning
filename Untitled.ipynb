```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2 + 3

print(X,Y)
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = 9
lr = 0.1
plt.plot(X,Y)
for i in range(30):
    grad = 2*(x-5)
    x = x - lr * grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(grad, x)

plt.show()
```
