```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#Data generation X,Y
def generateDataset(m):
    X = np.random.randn(m)*10 #std distribution
    noise = np.random.randn(m)
    y = 3 * X + 1 + 5*noise 
    return X,y
```

```python
X, y = generateDataset(100)
print(X.shape, y.shape)
print(X)
```

```python
def plotData(X,y, color="orange", title="Data"):
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.scatter(X,y,c=color)
    plt.show()
    
plotData(X,y)   
```

```python
#Normalization/Standardization of the data , ie: bringing the mean close to 0 and the std deviation equal to 1
def normaliseData(X):
    X = (X-X.mean())/X.std() #formula to obtain the normalized feature values
    return X
```

```python
plotData(X,y)
X = normaliseData(X)
plotData(X,y)
print(X.mean(), X.std())
```

```python
 def trainTestSplit(X, y, split=0.8):
        m = X.shape[0]
        
        #creating a new array 'data' to contain 2 columns, one for the X values and one for the y values so that when we apply 
        #shuffling operation, the corresponsing X and Y mapping should not get disturbed, ie: X and Y should shuffle together
        data = np.zeros((m,2))
        data[:, 0] = X
        data[:, 1] = y
        
        np.random.shuffle(data)
        
        split = int(m*split)
        
        XT = data[:split, 0]
        yT = data[:split, 1]
        
        Xt = data[split:, 0]
        yt = data[split: , 1]
        
        return XT, yT, Xt, yt
```

```python
XT, yT, Xt, yt = trainTestSplit(X,y)
```

```python
print(XT.shape, yT.shape)
print(Xt.shape, yt.shape)
```

```python
plt.scatter(XT, yT, color="orange", label="Train Data")
plt.scatter(Xt, yt, color="blue", label="Test Data")
plt.title("Train-Test Data")
plt.legend()
plt.show()
```
